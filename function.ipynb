```python
# WAP to create a calculater using function
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * /  ")
    
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif o == "/":
        print(div(a,b))
    else:
        print("invalid operator")
        
cal()
```

```python
# WAP to create a calculater using function
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

def cal(a,b,o):    
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif o == "/":
        print(div(a,b))
    else:
        print("invalid operator")

a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter + - * /  ")
cal(a,b,o)
```

```python
# WAP to create a calculater using function
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * /  ")
    
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif o == "/":
        print(div(a,b))
    else:
        print("invalid operator")
        
    x = input("Enter y for more cal.")
    if x == 'y':
        cal()
        
cal()
```
